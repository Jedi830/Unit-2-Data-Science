Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt

new_file = open("states.txt")
num_ny = 0
num_ca = 0
num_la = 0
num_tx = 0 
num_fl = 0
for x in new_file:
    if x == "NY\n":
        num_ny += 1
    elif x == 'CA\n':
        num_ca += 1
    elif x == 'LA\n':
        num_la += 1
    elif x == 'TX\n':
        num_tx += 1
    elif x == 'FL\n':
        num_fl += 1




height = [num_ny,num_ca,num_la,num_tx,num_fl]
x = ['NY','CA','LA','TX','FL']
plt.bar(x,height)
plt.xlabel("States")
plt.ylabel("Number of States repeated")
plt.title("Data science Lab")
````
